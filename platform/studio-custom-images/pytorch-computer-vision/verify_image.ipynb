## 4. Albumentations - Image Augmentation Test

```python
import albumentations as A
from albumentations.pytorch import ToTensorV2
import numpy as np

# Define a typical CV augmentation pipeline
transform = A.Compose([
    A.RandomCrop(width=224, height=224),
    A.HorizontalFlip(p=0.5),
    A.Normalize(mean=(0.485, 0.456, 0.406), std=(0.229, 0.224, 0.225)),
    ToTensorV2(),
])

# Create a dummy image and apply transforms
dummy_image = np.random.randint(0, 256, (256, 256, 3), dtype=np.uint8)
result = transform(image=dummy_image)
transformed_image = result["image"]

print(f"Input shape: {dummy_image.shape}")
print(f"Output shape: {transformed_image.shape}")
print(f"Output dtype: {transformed_image.dtype}")
print("\n✅ Albumentations augmentation pipeline works correctly")
```

## 5. TIMM - Pretrained Model Loading Test

```python
import timm
import torch

# List some available models
resnet_models = timm.list_models("resnet*", pretrained=True)[:5]
print(f"Sample available ResNet models: {resnet_models}")

# Create a model (without downloading pretrained weights to save time)
model = timm.create_model("resnet18", pretrained=False, num_classes=10)
model.eval()

# Run a forward pass
dummy_input = torch.randn(1, 3, 224, 224)
with torch.no_grad():
    output = model(dummy_input)

print(f"\nModel: resnet18 (num_classes=10)")
print(f"Input shape: {dummy_input.shape}")
print(f"Output shape: {output.shape}")
print(f"Total parameters: {sum(p.numel() for p in model.parameters()):,}")
print("\n✅ TIMM model creation and inference works correctly")
```

## 6. Ultralytics (YOLOv8) - Model Test

```python
from ultralytics import YOLO
import numpy as np

# Load YOLOv8 nano model (smallest, fastest to download)
model = YOLO("yolov8n.pt")

# Run inference on a dummy image
dummy_image = np.random.randint(0, 256, (640, 640, 3), dtype=np.uint8)
results = model(dummy_image, verbose=False)

print(f"Model: YOLOv8n")
print(f"Input shape: {dummy_image.shape}")
print(f"Detections: {len(results[0].boxes)}")
print(f"Classes available: {len(model.names)}")
print("\n✅ Ultralytics YOLOv8 inference works correctly")
```
